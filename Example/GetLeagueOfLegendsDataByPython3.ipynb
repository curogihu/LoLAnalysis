```python
import json
import urllib.request

urlData = "https://na.api.pvp.net/api/lol/na/v2.5/league/master?type=RANKED_SOLO_5x5&api_key=[APIKEY]"
webURL = urllib.request.urlopen(urlData)
data = webURL.read()
encoding = webURL.info().get_content_charset('utf-8')
summonerjson = json.loads(data.decode(encoding))

npSummonerjson = np.array(summonerjson["entries"])

tmp = []

for summoner in npSummonerjson:
    # print("id = " + summoner["playerOrTeamId"])
    #tmp.append((summoner["playerOrTeamId"], summoner["playerOrTeamName"]))
    tmp.append(summoner["playerOrTeamId"])

f = open('summoner-challengers-masters.csv', 'a', encoding="UTF-8") #ファイルが無ければ作る、の'a'を指定します

for summonerid in tmp:
    #f.write(summonername + "\n")
    # print(summonerid + "," + summonername)
    # print(summoner)
    # listData = []
    #listData.append(int(summoner))
    # csvWriter.writerow(summoner)
    f.write(summonerid + "\n")
f.close()
```

```python
url = "https://na.api.pvp.net/api/lol/na/v1.3/game/by-summoner/77720407/recent?api_key=[APIKEY]"

webURL = urllib.request.urlopen(url)
data = webURL.read()
encoding = webURL.info().get_content_charset('utf-8')
recentMatchJson = json.loads(data.decode(encoding))

npGameJson = np.array(recentMatchJson["games"])

matchDate = str(npGameJson[0]["createDate"])
matchId = str(npGameJson[0]["gameId"])
teamId = str(npGameJson[0]["teamId"])

if npGameJson[0]["stats"]['win']:
    winTeamId = teamId
    
else:
    if teamId == "100":
        winTeamId = "200"
    else:
        winTeamId = "100"
        
# matchDate + "," + matchId  + "," + teamId  + "," + winTeamId

fjson = open("Output/" + matchDate + "-" + matchId + ".json", "w")
json.dump(recentMatchJson, fjson, ensure_ascii=False, indent=4, sort_keys=True, separators=(',', ': '))

#変数1 = open(‘書き込みファイルのパス’, ‘w’)  #書き込むファイルを開く
#json.dump(変数2, 変数1) 
```
